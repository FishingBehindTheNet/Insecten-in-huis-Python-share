# 1.1.2 Statistieken per groep
Dit script is gebruikt om een inzicht te krijgen uit de data verzameld uit bijlage "1.1 Extractie rasterdata binnen buffers". Hierbij wordt de kolom "categorie" (de groepering van alle landgebruik typen) uit de kaart legenda geraadpleegd. Vervolgens wordt de data samengevat op basis van verschillende groepen en worden statistische gegevens weergegeven zoals: gemiddelde, standaard afwijking, standaard fout en sample grootte. De output bestaat uit een excel bestand met een tabblad voor elke groep aangegeven bij de "GroupBy" variabele.

```python
# Laad de benodigde ondersteunende bibliotheken in
from pathlib import Path
import pandas as pd
import numpy as np
import os
```

```python
# Vul hieronder het pad naar je Excel bestand in. 
ExelFile = Path(r"Output/TestInput_LGN2018_100.xlsx")

# Pad naa het bestand met begeleidend bestand wat over informatie beschikt van de legenda en andere groeperingen
KaartLegenda = Path(r"Input_data/List_LGN.xlsx")

# Variabelen die kunnen worden samengevat
GroupBy = ['Geslacht', 'Aantasters', 'Familie', 'Orde', 'Geslacht soort']
```

```python
# Laat de excel bestanden in
print("Excel bestanden inladen...")
InputDF = pd.read_excel(ExelFile, header=0)
KaartLegendaDF = pd.read_excel(KaartLegenda, header=0)
Categories = KaartLegendaDF['Categorie'].unique()

# Checkt of de output map bestaat en maakt deze anders
if not os.path.exists("Output"):
    os.makedirs("Output")


with pd.ExcelWriter(rf"Output\Stats_{ExelFile.stem}.xlsx") as writer:
    # Loopt over de categorieën
    for cat in GroupBy:
        # Maakt een lege df aan voor opslag
        OutputDF = pd.DataFrame()

        # Loopt over de categorieën uit de KaartLegenda
        for columns in Categories:
            column = f"C_{columns}"
            print(f"{columns} berekenen...")
            # Berekend verschillende basisstatistieken
            col_mean = InputDF.groupby(cat)[column].mean()
            col_std = InputDF.groupby(cat)[column].std()
            col_se = col_std / np.sqrt(InputDF.groupby(cat)[column].count())
            sample_size = InputDF.groupby(cat)[column].count()
            count_gt_zero = (InputDF[column] > 0).groupby(InputDF[cat]).sum()
            # Retrieve the unique values of 'Geslacht' for the current group
            Variable = InputDF.groupby(cat).groups.keys()

            # Create DataFrames for the current variable
            OutputIndividueel = pd.DataFrame({
                'groups' : Variable,
                'Land_type': columns,
                'Mean': col_mean,
                'SD': col_std,
                'SE': col_se,
                'n': sample_size,
                'Count > 0': count_gt_zero
            })

            print(f"{columns} opslaan...")
            OutputDF = pd.concat([OutputDF, OutputIndividueel])
        OutputDF.reset_index(inplace=True, drop=True)
        OutputDF.to_excel(writer, sheet_name=cat)

print(f"Output locaties: {os.getcwd()}\Output\Stats_{ExelFile.stem}.xlsx")
```
